```python
target_words_by_year = pd.read_csv('..\\output\\final\\target_words_by_year.csv')   # start importing the file
```

```python
target_words_by_year = target_words_by_year.drop(['Unnamed: 0'], axis=1)
target_words_by_year.head(5)                                         # check it again

```

```python
melted_words = pd.melt(target_words_by_year, 
                 id_vars=['set', 'words'], 
                 value_vars=['2001', '2002', '2003', '2004','2005', '2006','2007', '2008','2009', '2010',
                            '2011', '2012', '2013','2014', '2015','2016', '2017','2018', '2019', '2020', '2021'],
                      var_name = 'Year', 
                      value_name = 'Count')

melted_words.set_index('Year', inplace=True)
print(list(melted_words.columns))
```

## Plot some things



```python
set_1 = melted_words[melted_words.set == 1]               # data frame with only the rows matching each set

set_1.groupby('words')['Count'].plot(legend=True)

set_1.plot(x = 'Year', y = 'Count')
plt.legend(frameon=False)                                                        # 
plt.show()  
```
