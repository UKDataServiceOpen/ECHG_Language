```python
files = []
def import_results(input):
    for f in os.listdir(input):
        f = pd.read_csv(input + '\\'+ f)
        files.append(f)
    output = pd.concat(files)
    return output
```

```python
all_results = import_results("..\\results")
len(all_results)
```

```python
no_null_texts = all_results[~all_results['Text'].isnull()]
len(no_null_texts)
```
